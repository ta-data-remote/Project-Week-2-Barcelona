```python
pd.set_option('display.max_rows', 500)

accidents_7 = accidents_1.copy()
accidents_7["Intersection"] = accidents_7["Street"]
accidents_7["Highway"] = accidents_7["Street"]
accidents_7["Road"] = accidents_7["Street"]

accidents_7['Intersection'] = list(map(lambda x: 1 if ( " / " in x ) else 0, accidents_7['Intersection']))
accidents_7['Highway'] = list(map(lambda x: 1 if ( "-" in x or "(" in x) else 0, accidents_7['Highway']))
accidents_7['Road'] = list(map(lambda x: 0 if ( " / " in x or "-" in x or "(" in x) else 1, accidents_7['Road']))
```

```python
Serious_injuries_IHR = accidents_7.groupby("Serious injuries").sum()
Serious_injuries_IHR = Serious_injuries_IHR.drop(columns = ["Day","Hour", "Longitude", "Latitude", "Vehicles involved", "Mild injuries", "Victims"])
Serious_injuries_IHR
```

```python
Mild_injuries_IHR = accidents_7.groupby("Mild injuries").sum()
Mild_injuries_IHR = Mild_injuries_IHR.drop(columns = ["Day","Hour", "Longitude", "Latitude", "Vehicles involved", "Serious injuries", "Victims"])
Mild_injuries_IHR
```

```python
Vehicles_IHR = accidents_7.groupby("Vehicles involved").sum()
Vehicles_IHR = Vehicles_IHR.drop(columns = ["Day","Hour", "Longitude", "Latitude", "Mild injuries", "Serious injuries", "Victims"])
Vehicles_IHR
```
