```python
my_data = data[['Station', 'O3 Hour', 'O3 Quality', 'O3 Value']].copy()
my_data.columns = ['station', 'o3_hour', 'o3_quality', 'o3_value']
```

```python
o3_df = data[['Station', 'Air Quality', 'Longitude', 'Latitude', 'O3 Hour', 'O3 Quality', 'O3 Value', 'Generated', 'Date Time']].copy()
o3_df.columns = ['station', 'air_quality', 'longitude', 'latitude', 'o3_hour', 'o3_quality', 'o3_value', 'generated', 'date_time']
```

```python
nulls_o3 = o3_df[o3_df['o3_value'].isna() == True].index.tolist()
o3_clean = o3_df.drop(nulls_o3)
o3_clean.reset_index(drop=True,inplace=True)
o3_clean
```

```python
o3_clean.groupby('station').o3_value.agg(['min', 'max', 'mean', 'median']).round(2)
```

```python
o3_clean.groupby('o3_hour')['o3_value'].agg(['min','max','mean','median']).round(2)
```

```python
display(o3_clean.groupby(['station','o3_hour'])['o3_value'].agg(['min','max','mean','median']).round(2))
```

```python
monday = o3_clean[o3_clean['generated'].str.contains('05/11/2018|12/11/2018|19/11/2018|26/11/2018')]
tuesday = o3_clean[o3_clean['generated'].str.contains('06/11/2018|13/11/2018|20/11/2018|27/11/2018')]
wednesday = o3_clean[o3_clean['generated'].str.contains('07/11/2018|14/11/2018|21/11/2018|28/11/2018')]
thursday = o3_clean[o3_clean['generated'].str.contains('01/11/2018|08/11/2018|15/11/2018|22/11/2018|29/11/2018')]
friday = o3_clean[o3_clean['generated'].str.contains('02/11/2018|09/11/2018|16/11/2018|23/11/2018|30/11/2018')]
saturday = o3_clean[o3_clean['generated'].str.contains('03/11/2018|10/11/2018|17/11/2018|24/11/2018')]
sunday = o3_clean[o3_clean['generated'].str.contains('04/11/2018|11/11/2018|18/11/2018|25/11/2018')]
```

```python
mon_mean = monday['o3_value'].mean()
tue_mean = tuesday['o3_value'].mean()
wed_mean = wednesday['o3_value'].mean()
thu_mean = thursday['o3_value'].mean()
fri_mean = friday['o3_value'].mean()
sat_mean = saturday['o3_value'].mean()
sun_mean = sunday['o3_value'].mean()
```

```python
weekdays = [['Monday', mon_mean], ['Tuesday', tue_mean], ['Wednesday', wed_mean], ['Thursday', thu_mean], ['Friday', fri_mean], ['Saturday', sat_mean], ['Sunday', sun_mean]]

weekdf = pd.DataFrame(weekdays,columns=['Weekdays','O3 Mean']).set_index('Weekdays').round(2)
weekdf
```

```python
week_1 = o3_clean[o3_clean['generated'].str.contains('01/11/2018|02/11/2018|03/11/2018|04/11/2018|05/11/2018|06/11/2018|07/11/2018')]
week_2 = o3_clean[o3_clean['generated'].str.contains('08/11/2018|09/11/2018|10/11/2018|11/11/2018|12/11/2018|13/11/2018|14/11/2018')]
week_3 = o3_clean[o3_clean['generated'].str.contains('15/11/2018|16/11/2018|17/11/2018|18/11/2018|19/11/2018|20/11/2018|21/11/2018')]
week_4 = o3_clean[o3_clean['generated'].str.contains('22/11/2018|23/11/2018|24/11/2018|25/11/2018|26/11/2018|27/11/2018|28/11/2018')]
```

```python
w1_mean = week_1['o3_value'].mean()
w2_mean = week_2['o3_value'].mean()
w3_mean = week_3['o3_value'].mean()
w4_mean = week_4['o3_value'].mean()
```

```python
weeks = [['Week 1', w1_mean], ['Week 2', w2_mean], ['Week 3', w3_mean], ['Week 4', w4_mean]]

weeksdf = pd.DataFrame(weeks,columns=['Weeks','O3 Mean']).set_index('Weeks')
weeksdf
```
